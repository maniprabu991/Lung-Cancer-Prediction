```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Lung_Cancer_outlier_handled.csv')
```

```python
df.head(5)
```

```python
df['diagnosis_date'] = pd.to_datetime(df['diagnosis_date'])
df['end_treatment_date'] = pd.to_datetime(df['end_treatment_date'])


df['survival_days'] = (df['end_treatment_date'] - df['diagnosis_date']).dt.days


df.drop(columns=['diagnosis_date', 'end_treatment_date'], inplace=True)
```

```python
df.head(5)
```

```python
stage_mapping = {
    'Stage I': 1,
    'Stage II': 2,
    'Stage III': 3,
    'Stage IV': 4
}

df['cancer_stage'] = df['cancer_stage'].map(stage_mapping)
```

```python
df.head(5)
```

```python
category_mapping = {
    'smoking_status': {
        'Never Smoked': 0,
        'Passive Smoker': 0,
        'Former Smoker': 1,
        'Current Smoker': 1
    },
    'treatment_type': {
        'Chemotherapy': 0,
        'Radiation': 0,
        'Combined': 0,
        'Surgery': 1
    }
}

for col, mapping in category_mapping.items():
    df[col] = df[col].map(mapping)

df.rename(columns={
    'smoking_status': 'is_smoker',
    'treatment_type': 'is_surgical_treatment'
}, inplace=True)

```

```python
df.head(5)
```

```python
df[['is_smoker', 'is_surgical_treatment']].isnull().sum()
```

```python
df.columns
```

```python
#it replaces 'yes' with 1 and 'no' with 0.
df['family_history'] = df['family_history'].replace(['Yes','No'], [1,0])
```

```python
df.head(5)
```

```python
df.drop(columns=['country'], inplace=True)
```

```python
df = pd.get_dummies(
    df,
    columns=['gender'],
    drop_first=True
)
```

```python
df = df.astype(int)
```

```python
df.head(5)
```

```python
df.columns
```

```python
# Reduced feature set
important_cols = [
    'age',
    'cancer_stage',
    'is_smoker',
    'is_surgical_treatment',
    'other_cancer',
    'cirrhosis',
    'family_history',
    'bmi',
    'cholesterol_level',
    'hypertension',
    'asthma',
    'gender_Male'
]


X = df[important_cols]
y = df['survived']
```

```python
X
```

```python
X.columns
```

```python
y
```

```python
# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=12,
    class_weight='balanced_subsample',
    random_state=42,
    n_jobs=1
)

rf.fit(X_train, y_train)

```

```python
# Predict class labels
y_pred = rf.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Classification Report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, digits=4))
```

```python
print("No. of features model was trained on:", rf.n_features_in_)
print("Length of feature_importances_:", len(rf.feature_importances_))
```

```python
trained_features = [
    'age',
    'cancer_stage',
    'is_smoker',
    'is_surgical_treatment'
]
importance_df = pd.DataFrame({
    'feature': trained_features,
    'importance': rf.feature_importances_
}).sort_values(by='importance', ascending=False)

print(importance_df)
```
